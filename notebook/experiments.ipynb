# 2. Storing Document Vectors with FAISS

```python
# import Vector database
# FAISS is a vector database that allows us to store and query vectors efficiently.
# It is used to create a vector store from the documents and the embedding model.
# We will use the FAISS vector store to store the document vectors and query them later.

# FAISS is in memory, so it is fast and efficient for small to medium sized datasets.
# FAISS also support disk-based storage for larger datasets, but we will not use that here.

# Similar to FAISS Chroma is also a vector database that allows us to store and query vectors efficiently.
# Chroma is for a smaller dataset and is more lightweight compared to FAISS mostly used for prototyping and testing.


from langchain.vectorstores import FAISS
```

```python
docs[0].page_content
```

```python
embedding_model.embed_documents(docs[0].page_content)
```

```python
# Leng of current document
len(embedding_model.embed_documents(docs[0].page_content))
```

```python
# length of the embedding model
len(embedding_model.embed_documents(docs[0].page_content)[0])
```

```python
vectorstore = FAISS.from_documents(docs, embedding_model)
```

# 3. Retrieval

```python
# Now we can query the vector store to find the most similar documents to a given query.
# We can use the similarity_search method to find the most similar documents to a given query.
# We can specify the number of documents to return using the k parameter. top k documents will be returned.
# If k is not specified, it will return the top 3 documents by default.

query = "What is prompt engineering?"
vectorstore.similarity_search(query, k=3)
```

```python
vectorstore.similarity_search("Use variables in prompts", k=5)
```

```python
# you can use similarity search to find relevant documents for a given query without specifying k
relevant_docs = vectorstore.similarity_search("Use variables in prompts")
```

# 4. Generation Pipeline

```python
# Now we can create a retriever from the vector store.
# A retriever is an object that can be used to retrieve documents from the vector store.
# We can use the as_retriever method to create a retriever from the vector store
# The retriever can be used to retrieve documents from the vector store based on a given query
# The retriever will return the most relevant documents based on the query and the vector store.
retriever = vectorstore.as_retriever()
```

```python
retriever.invoke("What is prompt engineering?")
```

- Question - User query 
-  Context - Retrieving from vector database based on user query


```python
# create prompt template for the retriever

from langchain.prompts import PromptTemplate
```

```python
prompt_template = """
You are a helpful assistant. You will be given a context and a question.
Your task is to answer the question based on the context.
"If you don't have enough information, just say you don't have enough information."

Context: {context}

Question: {question}

Answer:

"""
```

```python
prompt = PromptTemplate(
    template=prompt_template,
    input_variables = ["context", "question"])
```

```python
prompt
```

- #### Final prompt

PromptTemplate(input_variables=['context', 'question'], input_types={}, partial_variables={}, template='\nYou are a helpful assistant. You will be given a context and a question.\nYour task is to answer the question based on the context.\n"If you don\'t have enough information, just say you don\'t have enough information."\n\nContext: {context}\n\nQuestion: {question}\n\nAnswer:\n\n')

```python
# LCEL - Langchain Expression Language
# import langchain_core.output_parsers

from langchain_core.output_parsers import StrOutputParser
```

```python
# create a parser object to parse the output of the prompt
parser = StrOutputParser()
```

```python
def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
# import RunablePassthrough
# runnablePassthrough is used to pass the input to the next step in the chain without any modification.
# it is used to pass the question to the prompt without any modification.

from langchain_core.runnables import RunnablePassthrough
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | parser
)
```

```python
rag_chain.invoke("what is prompt engineering?")
```
